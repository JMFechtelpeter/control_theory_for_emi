```python
from ctrl import utils

# data = utils.generate_random_dataset(10, 50)
data = utils.load_data()
```

Calculate the optimal control for $0.1 \leq \rho \leq 100$. Save the corresponding control energies, the squared deviation from the target value. Also calculate the average controllability.

```python
from tqdm import tqdm
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
from ctrl import discrete_optimal_control as doc
import custom_rcparams_paper as crc
import os

N = len(data)
n_feat = data[0]['X'].shape[1]

### Specify the values of rho (control cost). We test a log space from 0.1 to 100.
r_space = np.logspace(-1, 2, 50)

n_r = len(r_space)
traj_losses = np.zeros((n_r, N))
emp_traj_losses = np.zeros((N))
ctrl_energies = np.zeros((n_r, N, n_feat))
emp_ctrl_energies = np.zeros((N, n_feat))
ac = np.zeros((N, n_feat))
mc = np.zeros((N, n_feat))
B = np.eye(n_feat)      # Simulated inputs to the individual nodes
Q = np.eye(n_feat)      # Cost matrix of deviation from the reference trajectory

for j in tqdm(range(len(data))):  

    A = data[j]['A']
    X = data[j]['X']
    Yref = np.ones_like(X) * (-3)       # The reference trajectory is constantly -3 (the best possible EMA value)
    for r in range(n_r):
        R = np.eye(B.shape[1]) * r_space[r]                                     # Control input cost matrix
        U_opt = doc.tracking_optimal_control(A, B, Q, R, X, Yref)
        X_controlled = doc.step(A, B, X, U_opt)                                 # Trajectory with optimal control input
        traj_losses[r, j] = doc.seminorm(Q, (X_controlled - Yref).T).mean()     # Trajectory deviation
        ctrl_energies[r,j] = ((U_opt @ B.T) **2).mean(axis=0)                   # Control energy of optimal control
        ac[j] = doc.average_ctrb(A, 15)
        mc[j] = doc.modal_ctrb(A)
    emp_traj_losses[j] = ((X-Yref)**2).sum(axis=1).mean()                       # Empirical trajectory deviation
    emp_ctrl_energies[j] = ((data[j]['Inp'] @ data[j]['B'].T)**2).mean(axis=0)  # Empirical control energy


```

Plot mean trajectory deviation and control energy over $0.1 \leq \rho \leq 100$ 

```python
mean_traj = traj_losses.mean(axis=1)
sem_traj = stats.sem(traj_losses, axis=1)
mean_energy = np.mean(np.sum(ctrl_energies, axis=2), axis=1)
sem_energy = stats.sem(np.sum(ctrl_energies, axis=2), axis=1)

fig, ax = plt.subplots()
handles = []
line_t,  = ax.plot(r_space, mean_traj)
ax.fill_between(r_space, (mean_traj-sem_traj), (mean_traj+sem_traj),
                alpha=0.3)
line_e, = ax.plot(r_space, mean_energy, color=crc.colors.energy)
ax.fill_between(r_space, mean_energy-sem_energy, mean_energy+sem_energy, color=crc.colors.energy, alpha=0.3)
line_emp_t, = ax.plot(r_space, [emp_traj_losses.mean()]*n_r, linestyle='--')
line_emp_e, = ax.plot(r_space, [emp_ctrl_energies.mean()]*n_r, linestyle='--', color=crc.colors.energy)
ax.set(ylim=(-5,  100), ylabel='loss/energy', yticks=(0, 40, 80),
       xscale='log', xticks=(0.1, 1, 10, 100), xlabel=r'$\rho$')
empty_line, = ax.plot([0,0],[0,0],linestyle='none',marker='none')
fig.legend((line_t, line_e, empty_line), ['trajectory loss', 'control energy', '(dashed=empirical)'])
utils.set_axis_size(ax, 1.3, 0.95)
plt.savefig('figures/S4a.png')
```

Plot correlation of average controllability with node-wise control energy over $0.1 \leq \rho \leq 100$ 

```python
correls = np.array([[np.corrcoef((ctrl_energies[r, j], ac[j], mc[j])) 
                                 for j in range(len(data))] for r in range(n_r)])
fisher = np.arctanh(correls)        # To obtain mean and standard error, we Fisher transform the correlation coefficients
mean_correl_ac = fisher[:, :, 0, 1].mean(axis=1)
sem_correl_ac = stats.sem(fisher[:, :, 0, 1], axis=1)
mean_energy = np.mean(np.sum(ctrl_energies, axis=2), axis=1)
sem_energy = stats.sem(np.sum(ctrl_energies, axis=2), axis=1)
legend_labels = [r'ctrl energy $\times$ AC']

fig, ax = plt.subplots()
line_ac, = ax.plot(r_space, np.tanh(mean_correl_ac), color=crc.colors.ac)  # In the plots, we reverse the Fisher transform
ax.fill_between(r_space, np.tanh(mean_correl_ac-sem_correl_ac), np.tanh(mean_correl_ac+sem_correl_ac),
                color=crc.colors.ac, alpha=0.3)
ax.set(xscale='log', xticks=(0.1, 1, 10, 100), xlabel=r'$\rho$',
       ylabel='correlation', ylim=(-0.5, 1), yticks=(-0.4, 0, 0.4, 0.8), yticklabels=(r'$-$.4', r'.0', r'.4', r'.8'))
fig.legend((line_ac,), legend_labels)
utils.set_axis_size(ax, 1.3, 0.95)
fig.savefig('figures/S3b.png')
```

Plot control energy for each node ordered by average controllability (from high to low). Do this plot for $\rho=0.1$ and $\rho=100$, i.e. low and high constraints.

```python
fig, axes = plt.subplots(1,2, figsize=(2.6,1.1), layout='none')
AC_order = np.argsort(ac, axis=1)[:,::-1]
for r, title, ax in zip([0,-1], 
                        [rf'Low ctrl constraint ($\rho={r_space[0]}$)', 
                        fr'High ctrl constraint ($\rho={r_space[-1]}$)'], 
                        axes):
    meanwrtAC = np.take_along_axis(ctrl_energies[r], AC_order, axis=1).mean(axis=0)
    semwrtAC = stats.sem(np.take_along_axis(ctrl_energies[r], AC_order, axis=1), axis=0)
    line1, = ax.plot(range(n_feat), meanwrtAC, color=crc.colors.ac)
    ax.fill_between(range(n_feat), meanwrtAC - semwrtAC, meanwrtAC + semwrtAC, color=crc.colors.ac, alpha=0.3)
    ax.set_xticks(range(n_feat))
    ax.set_xticklabels(['high']+['']*(n_feat-2)+['low'])
    ax.set_ylabel('ctrl energy')
    ax.set_xlabel('AC')
    ax.set_title(title)
axes[0].set_yticks([0,4,8])
axes[1].set_yticks([0,0.03,0.06])
fig.legend([line1], ['Items ranked by AC'], loc='upper left', bbox_to_anchor=[0,0.5])
# fig.tight_layout()
utils.set_axis_size(axes[0], width=2.2, height=0.95)
plt.savefig('figures/S3c.png')
```
